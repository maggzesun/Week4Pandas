```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schools = school_data_complete['school_name'].unique()
Totals_Schools = len(schools)

students = school_data_complete['student_name'].unique()
Totals_Students = len(students)

Totals_Budget = school_data_complete['budget'].sum()

Average_Math_Score = round(student_data['math_score'].mean(),2)
Average_Reading_Score = round(student_data['reading_score'].mean(),2)

school_data_complete_math_passing = school_data_complete.loc[school_data_complete['math_score']>70]
school_data_complete_reading_passing = school_data_complete.loc[school_data_complete['reading_score']>70]
percentage_math_passing = round(100*len(school_data_complete_math_passing['math_score'])/len(school_data_complete['math_score']),2)
percentage_reading_passing = round(100*len(school_data_complete_reading_passing['reading_score'])/len(school_data_complete['reading_score']),2)

school_data_complete_passing = school_data_complete.loc[(school_data_complete['math_score']>70) & (school_data_complete['reading_score']>70)]
percentage_passing = round(100*len(school_data_complete_passing)/len(school_data_complete),2)

district_summary = pd.DataFrame({'Total Schools':[Totals_Schools], 'Total Students':[Totals_Students],'Total Budget':[Totals_Budget],
                                 'Average Math Score': [Average_Math_Score], 'Average Reading Score': [Average_Reading_Score],
                                 '% Passing Math': [percentage_math_passing], '% Passing Reading': [percentage_reading_passing],
                                 '% Students Passing': [percentage_passing]})

district_summary.dtypes
```

```python
district_summary['Total Students'] = district_summary['Total Students'].map("{:,}".format)
district_summary['Total Budget'] = district_summary['Total Budget'].map("{:,}".format)
district_summary.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary = school_data_complete.groupby('school_name')
Average_Math_Score = round(school_summary['math_score'].mean(),2)
Average_Reading_Score = round(school_summary['reading_score'].mean(),2)


Total_Budget = school_summary['budget'].sum()
Total_Students = school_summary.size()
Per_Student_Budget = Total_Budget/Total_Students

school_summary_math_passing = school_data_complete_math_passing.groupby('school_name')
percentage_summary_math_passing = round(100*school_summary_math_passing['math_score'].count()/school_summary['math_score'].count(),2)

school_summary_reading_passing = school_data_complete_reading_passing.groupby('school_name')
percentage_summary_reading_passing = round(100*school_summary_reading_passing['reading_score'].count()/school_summary['reading_score'].count(),2)

school_summary_passing = school_data_complete_passing.groupby('school_name')
percentage_summary_passing = round(100*school_summary_passing.size()/school_summary.size(),2)

school_summary_df = pd.DataFrame({'Total Students':Total_Students,'Total Budget':Total_Budget,'Per Student Budget':Per_Student_Budget,
                                 'Average Math Score': Average_Math_Score, 'Average Reading Score': Average_Reading_Score,
                                 '% Passing Math': percentage_summary_math_passing, '% Reading Math':percentage_summary_reading_passing,
                                 '% Overall Passing': percentage_summary_passing})

school_summary_merge = pd.merge(school_summary_df,school_data, on = 'school_name')

school_summary_merge_revised = school_summary_merge.drop(['size','budget'],axis = 1)

school_summary_merge_revised.rename(columns = {'type': 'Type'}, inplace = True)

school_summary_merge_revised = school_summary_merge_revised.set_index('school_name')
school_summary_merge_revised.index.name = 'School Name'
school_summary_merge_revised.head()
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
school_summary_merge_sorted_descending = school_summary_merge_revised.sort_values('% Overall Passing',ascending = False)

school_summary_merge_sorted_descending.index.name = 'School Name'
school_summary_merge_sorted_descending.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
school_summary_merge_sorted = school_summary_merge_revised.sort_values('% Overall Passing',ascending = True)

school_summary_merge_sorted.index.name = 'School Name'
school_summary_merge_sorted.head(5)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0,1000000,2000000,3000000,4000000]
group_names =['<$1m','$1m-2m','$2m-3m','$3m-4m']

school_summary_merge_revised['School Spending'] = pd.cut(school_summary_merge_revised['Per Student Budget'],bins,labels = group_names, include_lowest= True)

school_summary_merge_revised.index.name = 'school_name'
school_spending_merge = pd.merge(school_data_complete, school_summary_merge_revised, how="left", on=["school_name", "school_name"])

group_1 = school_spending_merge.loc[school_spending_merge['School Spending'] == '<$1m']
group_2 = school_spending_merge.loc[school_spending_merge['School Spending'] == '$1m-2m']
group_3 = school_spending_merge.loc[school_spending_merge['School Spending'] == '$2m-3m']
group_4 = school_spending_merge.loc[school_spending_merge['School Spending'] == '$3m-4m']

group_1_math_average = round(group_1['math_score'].mean(),2)
group_2_math_average = round(group_2['math_score'].mean(),2)
group_3_math_average = round(group_3['math_score'].mean(),2)
group_4_math_average = round(group_4['math_score'].mean(),2)

Average_Math_Score_Spending = [group_1_math_average,group_2_math_average,group_3_math_average,group_4_math_average]

group_1_reading_average = round(group_1['reading_score'].mean(),2)
group_2_reading_average = round(group_2['reading_score'].mean(),2)
group_3_reading_average = round(group_3['reading_score'].mean(),2)
group_4_reading_average = round(group_4['reading_score'].mean(),2)

Average_Reading_Score_Spending =[group_1_reading_average,group_2_reading_average,group_3_reading_average,group_4_reading_average]

group_1_math_pass_percentage = 100*group_1.loc[group_1['math_score']>70]['Student ID'].count()/group_1['Student ID'].count()
group_2_math_pass_percentage = 100*group_2.loc[group_2['math_score']>70]['Student ID'].count()/group_2['Student ID'].count()
group_3_math_pass_percentage = 100*group_3.loc[group_3['math_score']>70]['Student ID'].count()/group_3['Student ID'].count()
group_4_math_pass_percentage = 100*group_4.loc[group_4['math_score']>70]['Student ID'].count()/group_4['Student ID'].count()

math_pass_percentage = [group_1_math_pass_percentage,group_2_math_pass_percentage,group_3_math_pass_percentage,group_4_math_pass_percentage]

group_1_reading_pass_percentage = 100*group_1.loc[group_1['reading_score']>70]['Student ID'].count()/group_1['Student ID'].count()
group_2_reading_pass_percentage = 100*group_2.loc[group_2['reading_score']>70]['Student ID'].count()/group_2['Student ID'].count()
group_3_reading_pass_percentage = 100*group_3.loc[group_3['reading_score']>70]['Student ID'].count()/group_3['Student ID'].count()
group_4_reading_pass_percentage = 100*group_4.loc[group_4['reading_score']>70]['Student ID'].count()/group_4['Student ID'].count()

reading_pass_percentage = [group_1_reading_pass_percentage,group_2_reading_pass_percentage,group_3_reading_pass_percentage,group_4_reading_pass_percentage]

group_1_pass_percentage = 100*group_1.loc[(group_1['reading_score']>70) & (group_1['math_score']>70)]['Student ID'].count()/group_1['Student ID'].count()
group_2_pass_percentage = 100*group_2.loc[(group_2['reading_score']>70) & (group_2['math_score']>70)]['Student ID'].count()/group_2['Student ID'].count()
group_3_pass_percentage = 100*group_3.loc[(group_3['reading_score']>70) & (group_3['math_score']>70)]['Student ID'].count()/group_3['Student ID'].count()
group_4_pass_percentage = 100*group_4.loc[(group_4['reading_score']>70) & (group_4['math_score']>70)]['Student ID'].count()/group_4['Student ID'].count()

pass_percentage = [group_1_pass_percentage,group_2_pass_percentage,group_3_pass_percentage,group_4_pass_percentage]

school_summary_spending_df = pd.DataFrame({'School Spending': group_names, 'Average Math Score': Average_Math_Score_Spending,
                                          'Average Reading Score': Average_Reading_Score_Spending, '% Math Pass': math_pass_percentage,
                                          '% Reading Pass': reading_pass_percentage, '% Overall Pass': pass_percentage})

school_summary_spending_df
```
